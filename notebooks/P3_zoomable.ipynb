# Zoom and Customization of pathway visualization

A few examples, for Figure 4. 

```python
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
import json

sys.path.append("/Users/lish/li.github/Li_CVK/lcvk")
from polarPlot import *
from metModels import *
```

```python
# Get metabolic model
model = json.load(open('../data/metabolicModel_RECON3D_20210510.json'))
print(model['meta_data'])
model = json_convert_azmuth_mummichog(model)
pathwayslist = model['metabolic_pathways']
```

```python
P = pathwayslist[23]
P
```

```python
filename = clean_pathway_name(P['name'])
pathway_nodes, pathway_edges = [model['Compounds'][x] for x in P['cpds']], P['edges']
cplot_LCVK_pathway(pathway_nodes, pathway_edges, formula='charged_formula', cpd_name='id',
                    title='',  ydata='ratio_NOPS', 
                    width=4, height=4, fontsize=6,
                    rotation=30, show_names=False, yLabel_off=True,
                    outfile= '../output/Fig4_a_' +filename+'.pdf')
```

```python
# port 'Eicosanoid metabolism'
example_network_24 = {
    'edges': pathway_edges,
    'nodes': pathway_nodes
}
with open('../example_network_24.json', 'w', encoding='utf-8') as f:
    json.dump(example_network_24, f,  ensure_ascii=False, indent=2)  
```

```python
# cap max_H_C_allowed=1.6
cplot_LCVK_pathway(pathway_nodes, pathway_edges, formula='charged_formula', cpd_name='id',
                    title='',  ydata='ratio_NOPS', 
                    max_H_C_allowed=1.6, 
                    width=4, height=4, fontsize=6,
                    rotation=30, show_names=False, yLabel_off=True,
                    outfile= '../output/Fig4_b_' +filename+'.pdf')
```

```python
# cap max_H_C_allowed=1.6
cplot_LCVK_pathway(pathway_nodes, pathway_edges, formula='charged_formula', cpd_name='id',
                    title='',  ydata='ratio_NOPS', 
                    max_H_C_allowed=1.6, 
                    width=4, height=4, fontsize=6,
                    rotation=30, show_names=False, yLabel_off=True,
                    facecolors='k', linewidths=.2, edgecolors='k', s=56, alpha=1,
                    outfile= '../output/Fig4_notuse_' +filename+'.pdf')
```

```python
# bigger figure
cplot_LCVK_pathway(pathway_nodes, pathway_edges, formula='charged_formula', cpd_name='id',
                    title='',  ydata='ratio_NOPS', 
                    max_H_C_allowed=1.6, 
                    width=18, height=18, fontsize=6,
                    rotation=40, show_names=True, yLabel_off=True,
                    facecolors='blue', linewidths=.5, edgecolors='k', s=32, alpha=.6,
                    outfile= '../output/Fig4_c_named_' +filename+'.pdf')
```

```python
# with optimization
cplot_LCVK_pathway(pathway_nodes, pathway_edges, formula='charged_formula', cpd_name='id',
                    title='',  ydata='ratio_NOPS', 
                    max_H_C_allowed=1.6, 
                    width=18, height=18, fontsize=6,
                    rotation=40, show_names=False, yLabel_off=True,
                    optimization=True,
                      optimization_spacer=0.001, optimization_iter=100,
                    facecolors='blue', linewidths=.5, edgecolors='k', s=32, alpha=.6,
                    outfile= '../output/Fig4_d_' +filename+'.pdf')
```

```python
# with optimization, with names
cplot_LCVK_pathway(pathway_nodes, pathway_edges, formula='charged_formula', cpd_name='id',
                    title='',  ydata='ratio_NOPS', 
                    max_H_C_allowed=1.6, 
                    width=18, height=18, fontsize=6,
                    rotation=40, show_names=True, yLabel_off=True,
                    optimization=True,
                      optimization_spacer=0.001, optimization_iter=100,
                    facecolors='blue', linewidths=.5, edgecolors='k', s=32, alpha=.6,
                    outfile= '../output/Fig4_e_named_' +filename+'.pdf')
```
